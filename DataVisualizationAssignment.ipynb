```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")
p.line(x, y, line_width=2, color="blue", legend_label="sin(x)")
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
np.random.seed(42)
data = {
    'x': np.random.rand(50) * 10,
    'y': np.random.rand(50) * 10,
    'sizes': np.random.randint(10, 30, 50),     
    'colors': np.random.choice(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], 50)  # Random colors
}

df = pd.DataFrame(data)
source = ColumnDataSource(df)

p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label="X values", y_axis_label="Y values")

p.scatter(x='x', y='y', size='sizes', color='colors', alpha=0.6, source=source)

show(p)
```

```python
#3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20,25,30,35]
data = {'fruits': fruits, 'counts':counts}
source = ColumnDataSource(data=data)
p = figure(x_range=fruits, title="Fruit Counts", 
           toolbar_location=None, tools="", y_axis_label="Counts")
p.vbar(x='fruits', top='counts', width=0.6, source=source, 
       line_color='white', fill_color=factor_cmap('fruits', palette=Category10[4], factors=fruits))
output_notebook()
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Data", x_axis_label="Value", y_axis_label="Frequency")

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

output_notebook()
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar
from bokeh.models import LinearColorMapper, ColorBar
# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Set up the color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=np.min(data_heatmap), high=np.max(data_heatmap))

# Create the figure
p = figure(title="Heatmap", x_axis_label="X-axis", y_axis_label="Y-axis",
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()))

# Display the heatmap using 'image'
p.image(image=[data_heatmap], x=x.min(), y=y.min(), dw=x.max() - x.min(), dh=y.max() - y.min(),
        color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(p)
```
